```python
import sqlite3
from AnnoApp.pyutils import split_para
import json

def get_data():
    connection = sqlite3.connect('instance/AnnoApp.sqlite')
    cursor = connection.execute("SELECT id, title, body, annotation from post ")
    posts = []
    for row in cursor:
        posts.append(row)
    return posts

def format_data(post):
    id, para, annotation_string = post[0], post[2], post[3]
    para_words = split_para(para)
    annotation = json.loads(annotation_string)
    ner = ["O"] * len(para_words)
    for index, cls in annotation.items():
        ner[int(index)] = cls
    ml_data = {}
    ml_data["words"], ml_data["ner"], ml_data["id"] = para_words, ner, id
    return ml_data

def form_record_list(posts, max_length=0):
    record_list = []
    for post in posts:
        record = format_data(post)
        if max_length != 0:
            if len(record['words']) > max_length:
                record['words'] = record['words'][:max_length]
                record['ner'] = record['ner'][:max_length]
        record_list.append(record)
    return record_list
```
